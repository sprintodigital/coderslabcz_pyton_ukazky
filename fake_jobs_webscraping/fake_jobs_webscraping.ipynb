```python
import requests as re
from bs4 import BeautifulSoup as bs
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://realpython.github.io/fake-jobs/"
page = re.get(url)
print(page.text)
```

```python
results = bs(page.content, "html.parser")
print(results)
```

```python
jobs = results.find_all("div",class_="card")
print(len(jobs))
```

```python
for job in jobs:
    print(job.find(class_="title").text)
    print(job.find(class_="subtitle").text)
    print(job.find(class_="location").text.strip())
    print(job.find("time").text)
    
```

```python
rows = []
for job in jobs:
    row = {}
    row["title"] = job.find(class_="title").text
    row["subtitle"] = job.find(class_="subtitle").text
    row["city"] = job.find(class_="location").text.strip().split(",")[0]
    row["state"] = job.find(class_="location").text.strip().split(",")[1].strip()
    row["time"] = job.find("time").text
    rows.append(row)
    
df = pd.DataFrame(data = rows)
df["time"] = pd.to_datetime(df["time"])
display(df)
```
